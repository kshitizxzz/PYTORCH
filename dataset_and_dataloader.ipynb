```python
import torch
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
```

```python
X_train , X_test , y_train , y_test = train_test_split(df.iloc[: , 1:] , df.iloc[: , 0] , test_size=0.2)
```

```python
scaler = StandardScaler()
encoder = LabelEncoder()
```

```python
X_train_transform = scaler.fit_transform(X_train)
X_test_transform = scaler.transform(X_test)
```

```python
y_train_transform = encoder.fit_transform(y_train)
y_test_transform = encoder.transform(y_test)
```

```python
X_train_tensor = torch.from_numpy(X_train_transform).float()
y_train_tensor = torch.from_numpy(y_train_transform).float()

X_test_tensor  = torch.from_numpy(X_test_transform).float()
y_test_tensor  = torch.from_numpy(y_test_transform).float()
```

```python
from torch.utils.data import Dataset , DataLoader
```

```python
class CustomDataset(Dataset):
    def __init__(self , features , labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        return self.features.shape[0]
    
    def __getitem__(self , index):
        return self.features[index] , self.labels[index]
```

```python
train_dataset = CustomDataset(X_train_tensor , y_train_tensor)
test_dataset = CustomDataset(X_test_tensor ,  y_test_tensor )
```

```python
train_dataset[10]
```

```python
train_loader = DataLoader(train_dataset , batch_size = 32 , shuffle = True)
test_loader = DataLoader(test_dataset , batch_size = 32 , shuffle = True)
```

```python
import torch.nn as nn
```

```python
class MySimpleNN(nn.Module):
    def __init__(self , num_features):
        super().__init__()
        self.linear = nn.Linear(num_features , 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self ,features) :
        out = self.linear(features)
        out = self.sigmoid(out)

        return out 
```

```python
learning_rate = 0.1 
epochs = 25
```

```python
model = MySimpleNN(X_train_tensor.shape[1])
# create optimizer
optimizer = torch.optim.SGD(model.parameters() , lr = learning_rate )
# create loss function
loss_function = nn.BCELoss()
```

```python

# define loop
for epoch in range(epochs):

  for batch_features, batch_labels in train_loader:

    # forward pass
    y_pred = model(batch_features)

    # loss calculate
    loss = loss_function(y_pred, batch_labels.view(-1,1))

    # clear gradients
    optimizer.zero_grad()

    # backward pass
    loss.backward()

    # parameters update
    optimizer.step()

  # print loss in each epoch
  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```

```python
# Model evaluation using test_loader
model.eval()  # Set the model to evaluation mode
accuracy_list = []

with torch.no_grad():
    for batch_features, batch_labels in test_loader:
        # Forward pass
        y_pred = model(batch_features)
        y_pred = (y_pred > 0.8).float()  # Convert probabilities to binary predictions

        # Calculate accuracy for the current batch
        batch_accuracy = (y_pred.view(-1) == batch_labels).float().mean().item()
        accuracy_list.append(batch_accuracy)

# Calculate overall accuracy
overall_accuracy = sum(accuracy_list) / len(accuracy_list)
print(f'Accuracy: {overall_accuracy:.4f}')
```
