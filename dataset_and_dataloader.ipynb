```python
from sklearn.datasets import make_classification
```

```python
import torch
```

```python
# step1 Create a synthetic classification 
```

```python
X,y = make_classification(
    n_samples = 10,     # no. of samples
    n_features = 2 ,    
    n_informative = 2 ,   # no. of informative features
    n_redundant = 0,      # no. of redundant feaatures
    n_classes = 2,        # no. of classes
    random_state = 42     # for reproducibility
)
```

```python
X = torch.tensor(X , dtype=torch.float32)
```

```python
y = torch.tensor(y , dtype = torch.long)
```

```python
X
```

```python
y
```

```python
from torch.utils.data import Dataset , DataLoader
```

```python
class CustomDataset(Dataset):
    def __init__(self , features , labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        return self.features.shape[0]
    
    def __getitem__(self , index):
        return self.features[index] , self.labels[index]
```

```python
dataset = CustomDataset(X , y)
```

```python
len(dataset)
```

```python
dataset[2]
```

```python
dataloader = DataLoader(dataset , batch_size = 2 , shuffle=True)
```

```python
for batch_features , batch_loaders in dataloader:
    print(batch_features)
    print(batch_loaders)
    print('-'*50)
```
